```python
import io
d_string = io.StringIO("""
Nombre Mes Rate1 Rate2
Aira 1 12 23
Aira 2 18 73
Aira 3 19 45
Ben 1 53 19
Ben 2 22 87
Ben 3 19 45
Cat 1 22 87
Cat 2 67 43
Cat 3 45 32
""")
D = pd.read_table(d_string, sep=" ")
D
```

```python
Mmes = D.iloc[:, 1:4].groupby("Mes").agg("mean")
SDmes = D.iloc[:, 1:4].groupby("Mes").agg("std")
SDmes/Mmes
```

### Leer csv de una URl

```python
import requests

url = "http://www.stats.ox.ac.uk/pub/datasets/csb/ch4a.dat"
s = requests.get(url).content

c = pd.read_table(io.BytesIO(s), sep=r'\s+', engine='python', header = None)
c
```

```python
largo = len(c[0])
largo

fechas = pd.date_range('2000/1/1', periods=largo,  freq='M')
fechas
```

```python
ejes_mes = pd.date_range(fechas[0], fechas[largo-1], freq='M')
ejes_ano = pd.date_range(fechas[0], fechas[largo-1], freq='Y')
```

```python
serie = c[0]
serie.index=fechas
serie
```

```python
serie.plot(style='k.')
plt.show
```
